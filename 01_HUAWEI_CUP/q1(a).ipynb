```python
import pandas as pd
import numpy as np

```

```python
# 使用pandas的read_excel方法读取.xlsx文件
df1 = pd.read_excel('./竞赛发布数据/表1-患者列表及临床信息.xlsx', engine='openpyxl')
df2 = pd.read_excel('./竞赛发布数据/表2-患者影像信息血肿及水肿的体积及位置.xlsx', engine='openpyxl')
df3 = pd.read_excel('./竞赛发布数据/表3-患者影像信息血肿及水肿的形状及灰度分布.xlsx', engine='openpyxl')
df4 = pd.read_excel('./竞赛发布数据/表4-答案文件.xlsx', engine='openpyxl')
df5 = pd.read_excel('./竞赛发布数据/附表1-检索表格-流水号vs时间.xlsx', engine='openpyxl')
```

```python
# 从df1中提取"ID"和"发病到首次影像检查时间间隔"两列
subset_df1 = df1[["ID", " "]]

# 从df2中提取"ID", "首次检查流水号"和"HM_volume"三列
subset_df2 = df2[["ID", "HM_volume", "HM_volume1", "HM_volume2", "HM_volume3", "HM_volume4", "HM_volume5", "HM_volume6", "HM_volume7", "HM_volume8"]]

# 使用ID列合并两个数据子集
new_df = subset_df1.merge(subset_df2, on="ID")
new_df.rename(columns={'HM_volume': 'HM_volume0'}, inplace=True)


new_df
```

```python
subset_df3 = df5[["ID", "入院首次检查时间点",'随访1时间点','随访2时间点','随访3时间点','随访4时间点','随访5时间点','随访6时间点','随访7时间点','随访8时间点','随访9时间点','随访10时间点','随访11时间点','随访12时间点']]
subset_df3.rename(columns={"入院首次检查时间点": '随访0时间点'}, inplace=True)
subset_df3
```

```python
# import pandas as pd

# # 假设 df1 和 df2 已经被读取

# # 透视 df1 和 df2
df1_melted = new_df.melt(id_vars=["ID"], 
                      value_vars=[f"HM_volume{i}" for i in range(0, 9)],
                      var_name="Visit", value_name="HM_volume")

df2_melted = subset_df3.melt(id_vars=["ID"], 
                      value_vars=[f"随访{i}时间点" for i in range(0, 13)],
                      var_name="Visit", value_name="Date")

# # 对齐 Visit 列的名称
df1_melted["Visit"] = df1_melted["Visit"].str.replace("HM_volume", "随访").astype(str)
df2_melted["Visit"] = df2_melted["Visit"].str.replace("时间点", "").astype(str)

# 合并
merged_df = pd.merge(df1_melted, df2_melted, on=["ID", "Visit"])

# 过滤 NaN 值
new_df2 = merged_df.dropna(subset=["HM_volume"])
new_df2 = new_df2.sort_values(by=["ID",'Visit'])

new_df2
```

```python
import pandas as pd

# 假设你的DataFrame叫做df

# 划分数据
df0 = new_df2[new_df2['Visit'] == '随访0'].copy()
dfn = new_df2[new_df2['Visit'] != '随访0'].copy()

# 合并数据
dfn = dfn.merge(df0[['ID', 'Date', 'HM_volume']], on='ID', suffixes=('', '_0'))

# 计算时间差
dfn['Time_difference'] = (dfn['Date'] - dfn['Date_0']).dt.total_seconds() / 3600  # 转换为小时

# 计算体积差和百分比
dfn['Volume_difference'] = dfn['HM_volume'] - dfn['HM_volume_0']
dfn['Volume_percentage'] = (dfn['Volume_difference'] / dfn['HM_volume_0']) * 100


dfn
```

```python
dfn2 = dfn.merge(new_df[['ID', '发病到首次影像检查时间间隔']], on='ID', how='left')
dfn2['total_time'] = dfn2['Time_difference'] + dfn2['发病到首次影像检查时间间隔']
dfn2['是否发生水肿'] = np.where((dfn2['Volume_difference'] >= 6000) | (dfn2['Volume_percentage'] >= 33), 1, 0)

dfn2
```

```python
dfn2.to_excel('./output/output_temp1.xlsx', index=False, engine='openpyxl')
```

```python
# 筛选ID在sub001到sub100范围内的数据
df_filtered = dfn2[dfn2['ID'].isin([f'sub{str(i).zfill(3)}' for i in range(1, 101)])]

# 为每个ID计算所需值
results = []

for _, group in df_filtered.groupby('ID'):
    event_happened = ((group['是否发生水肿'] == 1) & (group['total_time'] <= 48)).any()
    if event_happened:
        min_time = group[group['是否发生水肿'] == 1]['total_time'].min()
    else:
        min_time = None
    results.append([group['ID'].iloc[0], int(event_happened), min_time])

# 创建新的DataFrame
df_new = pd.DataFrame(results, columns=['ID', '48小时内是否发生血肿扩张事件', '满足条件的最小total_time'])

df_new
```
